Aplikasi Menghitung Luas Persegi Panjang

```python
panjang = int(input("masukan panjang persegi panjang (cm): "))
lebar = int(input("masukan lebar persegi panjang (cm): "))

luas = panjang * lebar

print(f"luas persegi panjang yaitu {luas} cm^2.")
```

```python
pi = 22/7
jari_jari = float(input("Masukan jari jari bola"))

volume = 3/4 * pi * jari_jari**3

print (f"volume dengan jari jari {jari_jari} cm adalah {volume:.2f}")
```
